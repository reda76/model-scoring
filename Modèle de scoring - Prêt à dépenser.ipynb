```python
# Affiche un tableau des pourcentages des valeurs NaN pour chaque colonne
def tableau_pourcentage_quantité_NaN(dataframe):

    pourcentage_valeur_naan = pd.DataFrame(
        dataframe.isna().mean().round(4) * 100,
        columns=['Données manquantes en %'
                 ]).sort_values(by='Données manquantes en %', ascending=False)

    quantité_valeur_naan = pd.DataFrame(
        dataframe.isnull().sum(),
        columns=['Données manquantes en quantité'
                 ]).sort_values(by='Données manquantes en quantité',
                                ascending=False)

    # Fusion des deux dataframes
    pourcentage_quantité_valeur_naan = pd.concat(
        [pourcentage_valeur_naan, quantité_valeur_naan], axis=1)

    return (pourcentage_quantité_valeur_naan)
```

## Encodage des variables catégorielles

Étant donné que ces variables catégorielles sont des types objet, nous devons les encoder. En effet, les modèles de machines learning ne peuvent pas traiter les variables catégorielles.

Nous avons deux moyens pour cela, le premier serait de donner un nombre entier unique pour une catégorie, exemple : Masculin : 0 et Féminin : 1, cela s'appelle l'encodage d'étiquette.

Le second moyen serait de rendre les catégories en colonne toute ayant pour valeurs 0, mais 1 lorsque celle-ci possède la catégorie concernée. Voici un schéma plus compréhensible de ce deuxième moyen :

![Tableau encodage](images/tableau_encodage.png)

Ce second moyen s'appelle l'encodage à chaud.

Le premier moyen est intéressant lorsqu'il s'agit de deux valeurs uniques, comme dans notre exemple avec Masculin et Féminin.

Le second moyen est plus fiable, l'inconvénient est qu'il peut très vite augmenter considérablement le nombre de colonnes.

### Encodage d'étiquettes et à chaud

Comme expliqué précédemment, les variables catégorielles, de type **object** avec 2 catégories uniques, seront encodées par étiquette, et les variables avec plus de 2 catégories uniques, avec l'encodage à chaud.

Pour l'encodage des étiquettes, nous utiliserons Scikit-Learn **LabelEncoder** et pour l'encodage à chaud nous utiliserons **get_dummies(df)**, qui est une fonction pandas.

```python
# Encodage étiquette
label_encoder = LabelEncoder()
label_encoder_count = 0

# Itération des colonnes
for col in app_train:
    # Si la colonne est de type 'object'
    if app_train[col].dtype == 'object':
        # Si il y a deux valeurs uniques ou moins
        if len(list(app_train[col].unique())) <= 2:
            # Entrainement des données (avec 'fit')
            label_encoder.fit(app_train[col])
            # Transforme les données avec l'encodage
            app_train[col] = label_encoder.transform(app_train[col])
            app_test[col] = label_encoder.transform(app_test[col])

            # Nombre de colonnes encodées par étiquettes.
            label_encoder_count += 1

print(label_encoder_count, 'colonnes ont été encodées par étiquette.')
```

```python
# Encodage à chaud
app_train = pd.get_dummies(app_train)
app_test = pd.get_dummies(app_test)

print("Format de la dataframe d'entrainement: ", app_train.shape)
print('Format de la dataframe de test: ', app_test.shape)
```

### Alignement des données d'entrainement et de test

Nous devons avoir les mêmes colonnes dans les données d'entrainement et de test. L'encodage à chaud a créé plus de colonnes dans les données d'entrainement, car il y avait des variables catégorielles avec des catégories non représentées dans les données de test. 

Pour supprimer les colonnes des données d'entrainement qui ne figurent pas dans les données de test, nous avons besoin d''**align**' les dataframe, **align** permets d'aligner les données des deux dataframe. 

Nous extrayons la colonne target des données d'entrainement (elle n'est pas dans les données de test mais nous devons la conserver). Lorsque nous effectuons l'alignement, nous devons nous assurer de définir **axis = 1** pour aligner les dataframes en fonction des colonnes et non des lignes. 

```python
train_labels = app_train['TARGET']

# Alignement des données de formation et de test, seul les colonnes
# présentes dans les deux dataframe seront gardées.
app_train, app_test = app_train.align(app_test, join='inner', axis=1)

# Nous recréons ensuite la colonne 'target', en y replaçant ses valeurs que
# nous avons stockées dans la variable 'train_labels'
app_train['TARGET'] = train_labels

print("Format de la dataframe d'entrainement: ", app_train.shape)
print('Format de la dataframe de test: ', app_test.shape)
```

>Le `join='inner'`, se retrouve aussi dans le langage SQL avec la commande INNER JOIN, c'est un type de jointures très communes pour lier plusieurs tables entre-elles, voici une illustration pour mieux comprendre comme les colonnes des deux dataframes sont alignées.

>![Inner Join](images/sql-ensemble-intersect-300.png)

Les ensembles de données de la dataframe d'entrainement et de test ont désormais les mêmes fonctionnalités que celles requises pour l'apprentissage automatique. Le nombre de colonnes a considérablement augmenté en raison de l'encodage à chaud. À un moment donné, nous voudrons probablement essayer de réduire le nombre de colonnes (en supprimant les colonnes qui ne sont pas pertinentes) pour réduire la taille des ensembles de données.

# Corrélations

Une façon d'essayer de comprendre les données consiste à rechercher des corrélations entre les variables et la cible. Nous pouvons calculer le coefficient de corrélation de Pearson entre chaque variable et la cible en utilisant la méthode **'.corr'**.

Une corrélation négative entre deux variables indique que les valeurs d'une variable tendent à augmenter lorsque celles de l'autre variable diminuent.

Le coefficient de corrélation n'est pas la meilleure méthode pour représenter la "pertinence" d'une variable, mais il nous donne une idée des relations possibles au sein des données. Voici quelques interprétations générales de la valeur absolue du coefficient de corrélation :

- 0.00-0.19 « très faible »

- 0.20-0.39 « faible »

- 0.40-0.59 « modéré »

- 0.60-0.79 « forte »

Une méthode simple de construction de caractéristiques est appelée **caractéristiques polynomiales**. Dans cette méthode, nous créons des caractéristiques qui sont des puissances de caractéristiques existantes, ainsi que des termes d'interaction entre les caractéristiques existantes.  Le but est de regarder si la fusion de deux variables peut donner une nouvelle variable pertinente.  Prenons pour exemple l'agriculture de manière vulgarisée, si nous avons un champ, et que nous souhaitons améliorer le rendement de notre culture. Nous avons comme moyen l'engrais et l'eau, mais que se passe-t-il si nous fusionnons l'eau et l'engrais ? Que se passe-t-il si nous fusionnons l'eau avec l'eau, qui serait en fait, ajoutée plus d'eau ? Voici vulgairement ce que sont les caractéristiques polynomiales et à quoi elles servent, où l'ont peut capturer de potentielles interactions entre les variables, avec des variables déjà existantes.

Par exemple, nous pouvons créer des variables **'EXT_SOURCE_1^2'** et **'EXT_SOURCE_2^2'** ainsi que des variables telles que **'EXT_SOURCE_1 '** x **'EXT_SOURCE_2'**, **'EXT_SOURCE_1'** x **'EXT_SOURCE_2^2'**, **'EXT_SOURCE_1^2'** x **'EXT_SOURCE_2^2'**, et ainsi de suite. Ces caractéristiques, qui sont une combinaison de plusieurs variables individuelles, sont appelées **termes d'interaction** car elles capturent les interactions entre les variables. 

En d'autres termes, alors que deux variables en elles-mêmes peuvent ne pas avoir une forte influence sur la cible (**TARGET**), leur combinaison en une seule variable d'interaction peut montrer une relation avec la cible. 

Les termes d'interaction sont couramment utilisés dans les modèles statistiques pour capturer les effets de plusieurs variables. Nous pouvons en essayer quelques-uns pour voir s'ils peuvent aider notre modèle à prédire si un client remboursera ou non un prêt.

Cela peut aussi être utilisé pour voir si nous avons trop de données et avons atteint un plafond.

Ça peut permettre de réduire la quantité de données utilisées et ainsi gagner du temps sur notre notebook. En entreprise les données peuvent parfois couter cher, cette technique permet alors aussi de réduire les couts.

```python
# Cette fonction nous sera utile pour savoir si notre quantité de données
# est suffisante pour chaque modèle
def courbe_apprentissage(modele):
    # Nous reprenons la variable 'model' où est contenue le meilleur modèle
    # train_sizes : quantité de données à utiliser pour l'entrainement (en %)
    N, train_score, val_score = learning_curve(modele,
                                               X_train,
                                               y_train,
                                               train_sizes=np.linspace(
                                                   0.1, 1.0, 10),
                                               cv=5)

    # Chaque portion (le troisième élément du np.linspace est la quantité 
    # de portion) se retrouve dans la variable N.
    print(N)
    %matplotlib inline
    plt.plot(N, train_score.mean(axis=1), label="train")
    plt.plot(N, val_score.mean(axis=1), label="validation")
    plt.xlabel('train_sizes')
    plt.legend()
```

```python
%%time
courbe_apprentissage(KNeighborsClassifier(n_neighbors=2))
```

Nous avons donc ici 10 portions du X_train et y_train par tranche de 10%, et ensuite notre graphique.

On constate sur notre graphique qu'il aurait fallu plus de données pour que notre modèle soit entrainé au maximum, mais nous pouvons quand même peut être avoir un bon score, nous utiliserons le maximum de données, donc la totalité de notre jeu de donnée.

>En somme, l'objectif est d'avoir une courbe de validation qui converge vers la courbe d'entraînement avec des scores élevés pour les deux ensembles, ce qui suggère que le modèle généralise bien sans être trop ajusté aux données d'entraînement.

## Amélioration du modèle

Si nous améliorons notre modèle seulement via les données de test, notre modèle sera réglé dessus et nous risquons de perdre en fiabilité sur le score final.

Nous allons donc découper une troisième partie dans notre data, appelé **"validation set"**.

Nous allons donc améliorer notre score en réglant les paramètres avec les données d'entrainement sur les données de test, une fois le modèle réglé, nous validerons le score avec la data de validation.

Nous allons utiliser plusieurs morceaux pour valider notre modèle pour rendre le plus fiable possible son score.

Pour cela, nous allons faire une **Cross Validation**.

Voici une illustration, ce sera surement plus clair comme ça !

![Cross Validation](images/cross_validation.png)

De cette manière, nous pourrons par la suite comparer précisément plusieurs modèles, et choisir le bon modèle avec une plus grande fiabilité, car nous aurons leurs scores moyens.

Nous utiliserons ici la métrique **roc auc** qui pénalise les faux négatifs, plus de détails à ce sujet sont donnés un peu plus loin dans le notebook dans la section de la matrice de confusion.

```python
def cross_validation(model):
    score = cross_val_score(model, X_train, y_train, cv=5,
                            scoring='roc_auc').mean()

    return score
```

```python
cross_validation(KNeighborsClassifier())
```

Nous voyons ici que lorsque que nous évaluons notre modèle avec un validation set, nous avons un score bien différent puisque les données utilisées pour l'évaluer ne sont pas réglées sur nos données d'entrainement.

On se retrouve alors avec un score bien plus faible ! Il est cependant plus proche de la réalité lorsqu'il faudra prédire si une personne paiera son prêt ou non, car ce seront des nouvelles données non réglées sur les données d'entrainement.

### Validation curve

Nous pouvons aller encore plus loin, en modifiant les paramètres du classifier grâce à la **validation curve**.

La validation curve va nous permettre de trouver avec quel nombre de **k** notre modèle aura le meilleur score.

```python
%%time
# Choix du model
model = KNeighborsClassifier()

# param_name='n_neighbors' : nom de l'hyperparamètre que l'ont souhaite régler
# param_range=np.arange(1, 20) : Nombre de k voisin, nous allons tester chaque quantité
# pour avoir la quantité qui nous donnera le meilleur score
train_score, val_score = validation_curve(model,
                                          X_train,
                                          y_train,
                                          param_name='n_neighbors',
                                          param_range=np.arange(1, 10),
                                          cv=5)
```

Regardons le résultat à l'aide d'un graphique.

### Matrice de confusion

Faisons maintenant une matrice de confusion avec la fonction **confusion_matrix**.

Elle permet de vérifier le nombre de faux positifs, de faux négatifs, de vrais positifs et de vrais négatifs.

Voici une illustration.

![Matrice](images/matriceconfusion.png)

- Si le modèle prédit un **positif**, et que c'est bien un **positif**, alors c'est un **vrai positif**.
- Si le modèle prédit un **positif**, alors que c'est un **négatif**, alors c'est un **faux positif**.

- Si le modèle prédit un **négatif**, alors que c'est un **positif**, alors c'est un **faux négatif**.
- Si le modèle prédit un **négatif**, et que c'est bien un **négatif**, alors c'est un **vrai négatif**.

```python
%%time
confusion_matrix(y_test, model.predict(X_test))
```

Cela nous permet de savoir si le modèle ne fait pas une même prédiction pour toutes les données.

Voici un exemple, imaginons que nous avons un modèle qui doit prédire si une voiture sort du tunnel, ou une moto.

Disons que le modèle prédit à chaque fois une voiture, et que 80% des sorties du tunnel sont des voitures.

Alors notre modèle aura 80% de réussite, mais vous voyez bien qu'il y a un problème, puisqu'il n'a fait que prédire la même chose. Sur un autre tunnel, il suffit qu'il n'y ai que 30% de voiture et le score du modèle sera alors de 30%. 

C'est pour cela que nous avons utilisé comme métrique le **roc auc**, qui pénalise les faux positifs.

## Ensemble learning

L'Ensemble learning est une technique qui consiste à entrainer plusieurs modèles de machine learning pour ensuite considérer l'ensemble de leurs prédictions.

Pour ça il existe trois grandes méthodes : 
- **le bagging** 
- **le boosting**
- **le stacking**

Les algorithmes qui repose sur ses méthodes, comme l'algorithme de **Random Forest**, compte parmi les plus performants dans le monde du machine learning.

Nous pensons qu'il serait bien de terminer par cela afin de potentiellement avoir le score le plus élevé/fiable ! Il peut y avoir une variance un peu élevée sur les modèles précédents selon les parties de données utilisées pour les entrainer.

L'Ensemble learning permettras donc également d'avoir un score final avec une variance beaucoup plus faible.

En utilisant la moyenne de plusieurs modèles, nous pouvons alors en tirer un meilleur score, voici une petite illustration pour comprendre l'esprit de l'Ensemble Learning.

![ensemble_learning](images/ensemble_learning.png)

Nous pouvons donc même avoir plusieurs modèles ayant chacun un score de 55%, les combiner, et obtenir un score se rapprochant des 100%.

Attention cependant, **aucun modèle de doit avoir un score inférieur à 50%**, sinon le score des modèles combinés tendra vers les **0%** plutôt que les 100%.

Les modèles doivent aussi ne pas avoir les mêmes mauvaises prédictions ! L'idée est qu'une mauvaise prédiction soit ajustée par les bonnes prédictions des autres modèles.

Laissez-nous donc vous expliquer les différentes parties qui sont, **le bagging**, **le boosting** et **le stacking**.

Cela aidera à la compréhension du code par la suite.

#### Le Bagging

L'idée ici est de créer plusieurs entités d'un même modèle, et d'entrainer chacune de ses entités sur une portion aléatoire de notre dataset, nous avons donc avec ce procédé des résultats diversifiés puisqu'ils n'ont pas été entrainés sur les mêmes données.

Voici une illustration, c'est toujours utile !

![bagging](images/bagging.png)

L'exemple de **bagging** le plus connu est le **Random Forest**.

#### Le Boosting

L'idée est d'entrainer l'un après l'autre plusieurs modèles relativement faibles en demandant à chaque modèle d'essayer de corriger les erreurs effectuées par son prédécesseur.

Du coup on obtient un ensemble de modèles complémentaires dans lequel les faiblesses des uns sont compensées par les forces d'autre.

La différence avec le **Bagging**, car c'est très ressemblant, est que le **Boosting** utilise des modèles à score faible, des modèles sous entrainés.

Pour faire cela, il existe deux grands algorithmes qui sont **AdaBoost** et **Gradient Boosting**.

#### Le Stacking

L'idée du **Stacking** est d'entrainer un modèle de machine learning par-dessus les prédictions de plusieurs autres modèles.

C'est-à-dire qu’au lieu de simplement rassembler les résultats de nos modèles pour retenir une prédiction majoritaire, comme avec le **Bagging** et le **Boosting**, on demande à un modèle d'apprendre à reconnaitre qui a tort et qui a raison parmi les modèles pour lui-même prédire le résultat final.

![stacking](images/stacking.png)

Commençons, **scikit-learn** permet de faire cela facilement avec la fonction **Voting Classifier**.

Nous allons ici regarder quelles sont les variables les plus importantes que notre modèle a utilisées pour prédire si un client remboursera son prêt ou non.

Et SHAP pour avoir une interprétation locale.

```python
def importance_variable(model):
    features = app_train.columns
    importances = model.feature_importances_
    indices = np.argsort(importances)

    num_features = 10

    plt.title('Importance des variables')

    # Ne trace que le nombre de caractéristiques défini dans la variable 
    # num_features (15)
    plt.barh(range(num_features),
             importances[indices[-num_features:]],
             color='b',
             align='center')

    plt.yticks(range(num_features), [features[i] for i in indices[-num_features:]])

    plt.xlabel('Importance')
    plt.show()
```
